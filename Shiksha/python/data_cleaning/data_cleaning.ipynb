```python
df["Order Date"]=pd.to_datetime(df["Order Date"],errors='coerce')
df["Order Date"].isnull().sum()
```

```python
df["Ship Date"]=pd.to_datetime(df["Ship Date"],errors='coerce')
df["Ship Date"].isnull().sum()
print(df[["Order Date","Ship Date"]])
```

```python
#Profit Margin
df["Profit Margin"]=df["Profit"]/df["Sales"]

```

```python
#Loss Flag
df["Loss Flag"]=df["Profit"]<0
```

```python
#Year 
df["Year"] = df["Order Date"].dt.year
print(df[["Order Date","Year"]].head())
```

```python
#Month 
df["Month"] = df["Order Date"].dt.month
print(df[["Order Date","Month"]].head())
```

```python

#Total Sales
df["total_sales"]=df["Sales"].sum()
```

```python
#Total Profit 
df["Total Profit"]=df["Profit"].sum()
```

```python


# Sales By Category 
print(df.groupby("Category")["Sales"].sum())

# Profit
(df.groupby("Region")["Profit"].sum())


```

```python
df.to_csv("Clean_superstore.csv" , index=False)
```
