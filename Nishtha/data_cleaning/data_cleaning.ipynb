```python
df["Order Date"] = pd.to_datetime(df["Order Date"], errors="coerce") # Converts Order Date text to standard Date format.
df["Ship Date"] = pd.to_datetime(df["Ship Date"], errors="coerce")   # Converts Ship Date text to standard Date format.
print(df[["Order Date", "Ship Date"]]) # Displays only the specific date columns to verify conversion.
```

```python
# Profit Margin
df["Profit Margin"] = df["Profit"] / df["Sales"]

#Loss Flag
df["Loss Flag"] = df["Profit"]<0

#Year Column
df["Year"] = df["Order Date"].dt.year
print(df[["Order Date","Year"]].head())

#Month Column
df["Month"] = df["Order Date"].dt.month
print(df[["Order Date","Month"]].head())
```
